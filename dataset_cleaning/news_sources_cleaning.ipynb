### News Sources:

We retrieved this dataset of reputable news sources that are authentic and known to publish reliable news reports. However, after further evaluation we excluded this dataset from the ontology, as the Veracity architecture already has other features that checks the credibility of the news publisher of the message parsed into the algorithm. The purpose of the ontology is to ensure the news message matches factual information and is within the domain of known disaster events and concepts, hence the reason we discarded the use of this dataset.
